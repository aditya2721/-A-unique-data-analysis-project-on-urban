# Swiggy Delivery Tracking Project

This notebook aims to build a system for tracking Swiggy deliveries. This could involve visualizing delivery routes, analyzing delivery times, identifying bottlenecks, and optimizing delivery processes.

To begin, we'll set up the environment and define the initial steps.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# You might also need libraries for geographical plotting, e.g., folium or plotly.express
# import folium
# import plotly.express as px
```

```python
from datetime import datetime, timedelta
import random

def generate_synthetic_delivery_data(num_deliveries=100):
    data = []
    start_time = datetime(2023, 10, 26, 8, 0, 0)

    for i in range(num_deliveries):
        order_id = f"ORD-{1000 + i}"
        delivery_id = f"DEL-{2000 + i}"

        # Simulate order placement and dispatch
        order_placed = start_time + timedelta(minutes=random.randint(0, 1440))
        restaurant_pickup = order_placed + timedelta(minutes=random.randint(10, 30))
        rider_pickup = restaurant_pickup + timedelta(minutes=random.randint(2, 10))
        delivery_attempt = rider_pickup + timedelta(minutes=random.randint(15, 60))

        # Simulate coordinates (e.g., around a city like Bengaluru)
        pickup_lat = round(12.9 + random.uniform(-0.05, 0.05), 6)
        pickup_lon = round(77.5 + random.uniform(-0.05, 0.05), 6)
        delivery_lat = round(12.9 + random.uniform(-0.1, 0.1), 6)
        delivery_lon = round(77.5 + random.uniform(-0.1, 0.1), 6)

        # Simulate delivery status
        statuses = ['Pending', 'Picked Up', 'Delivered', 'Cancelled']
        delivery_status = random.choice(statuses)

        # Adjust delivery_attempt based on status for realism
        if delivery_status == 'Delivered':
            delivery_actual = delivery_attempt + timedelta(minutes=random.randint(0, 10))
        elif delivery_status == 'Cancelled':
            delivery_actual = random.choice([order_placed, rider_pickup]) # Cancelled before actual delivery
        else: # Pending, Picked Up
            delivery_actual = None # Still in progress

        data.append({
            'order_id': order_id,
            'delivery_id': delivery_id,
            'order_placed_at': order_placed,
            'restaurant_pickup_at': restaurant_pickup,
            'rider_pickup_at': rider_pickup,
            'delivery_attempted_at': delivery_attempt,
            'delivery_actual_at': delivery_actual,
            'pickup_latitude': pickup_lat,
            'pickup_longitude': pickup_lon,
            'delivery_latitude': delivery_lat,
            'delivery_longitude': delivery_lon,
            'delivery_status': delivery_status,
            'estimated_delivery_time_minutes': random.randint(30, 75)
        })

    df_deliveries = pd.DataFrame(data)

    # Calculate delivery duration if delivered
    df_deliveries['delivery_duration_minutes'] = (df_deliveries['delivery_actual_at'] - df_deliveries['rider_pickup_at']).dt.total_seconds() / 60
    df_deliveries.loc[df_deliveries['delivery_status'] != 'Delivered', 'delivery_duration_minutes'] = np.nan

    return df_deliveries

synthetic_delivery_df = generate_synthetic_delivery_data(num_deliveries=200)

print(f"Generated {len(synthetic_delivery_df)} synthetic delivery records.")
display(synthetic_delivery_df.head())
```

```python
# Calculate the percentage of deliveries by status
delivery_status_percentage = synthetic_delivery_df['delivery_status'].value_counts(normalize=True) * 100

print("Percentage of Deliveries by Status:")
display(delivery_status_percentage)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=delivery_status_percentage.index, y=delivery_status_percentage.values, palette='viridis')
plt.title('Percentage Distribution of Delivery Statuses')
plt.xlabel('Delivery Status')
plt.ylabel('Percentage (%)')
plt.ylim(0, 100) # Ensure y-axis goes from 0 to 100
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add percentage labels on top of the bars
for index, value in enumerate(delivery_status_percentage.values):
    plt.text(index, value + 1, f'{value:.1f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
